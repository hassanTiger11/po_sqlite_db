```python
'''
In the following I will edit rgb table
I will drop all the non relevant columns like:
    treatment
    plot 
    genotype 
I will keep the plant_name so it can be a foriegn key to field_book
That way we can lose redundant data and we can get this data from the 
field_book.

One conclusion I got from field book:
      {treatment, plot, genotype} hasn't changed overtime for each scan
'''
import sqlite3
import pandas as pd
from IPython.display import display

# global values
po9 = sqlite3.connect('po9.db')
po10 = sqlite3.connect('po10.db')
flir = pd.read_sql_query('SELECT * FROM flir', po9)
```

```python
#inspect 
display(flir)
print(flir.columns)
```

```python
'''
For the flir data I am not sure which columns makes it unique
'''

dups = flir[flir.duplicated(keep=False)]
display(dups)

```

```python
'''
ops:
1) drop columns
2) set primary keys
'''
#1) drop columns

drop_unwanted_columns = flir.drop(columns = ['level_0', 'index'])
dups = drop_unwanted_columns[drop_unwanted_columns.duplicated(keep=False)]
display(dups)
```

```python
'''
From the data we can see tha each plant has one scan per date
let's check that out
'''
#check that date+name is not redundant
name_date = drop_unwanted_columns['plant_name'] + drop_unwanted_columns['date']
+drop_unwanted_columns['lon'] + drop_unwanted_columns['lat']
print(f'len(name_date): {len(name_date)}')
print(f'len(set(name_date)): {len(set(name_date))}')
if(len(name_date) == len(set(name_date))):
    print('equal')
#Thus, the combination of date name is unique
```

```python
#identify primary keys
print(drop_unwanted_columns.columns)
is_dup_df = drop_unwanted_columns.duplicated(subset=['plant_name', 'date','bounding_area_m2', 'quartile_3'],keep=False)
dups = drop_unwanted_columns[is_dup_df]
display(dups)
```

```python
'''
commit new table drop_unwanted_columns
then edit with the script in the next cell
'''
schema = '''
CREATE TABLE IF NOT EXISTS "flir" (
"index" INTEGER,
  "date" TEXT,
  "treatment" TEXT,
  "plot" TEXT,
  "genotype" TEXT,
  "lon" REAL,
  "lat" REAL,
  "min_x" INTEGER,
  "max_x" INTEGER,
  "min_y" INTEGER,
  "max_y" INTEGER,
  "nw_lat" REAL,
  "nw_lon" REAL,
  "se_lat" REAL,
  "se_lon" REAL,
  "bounding_area_m2" REAL,
  "roi_temp" REAL,
  "quartile_1" REAL,
  "mean" REAL,
  "median" REAL,
  "quartile_3" REAL,
  "variance" REAL,
  "std_dev" REAL,
  "labels" INTEGER,
  "plant_name" TEXT,
  "double_lettuce" REAL,
  PRIMARY KEY('plant_name', 'date','bounding_area_m2', 'quartile_3' )
);
'''

drop_unwanted_columns.to_sql('flir', po10, if_exists='replace')
```

```python
table = 'flir'
sql_command = f'''

PRAGMA foreign_keys=off;

BEGIN TRANSACTION;

ALTER TABLE {table} RENAME TO old_{table};

-- define the primary key constraint here
{schema}

INSERT INTO {table} SELECT * FROM old_{table};

DROP TABLE old_{table};

COMMIT;

PRAGMA foreign_keys=on;


'''
cursor = po10.cursor()
cursor.executescript(sql_command)
```
