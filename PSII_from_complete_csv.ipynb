```python
import pandas as pd
from IPython.display import display

psii_complete = pd.read_csv('season10_lettuce_complete_data.csv')
display(psii_complete)

```

```python
import os
from IPython.display import display
def get_csv_files_paths_and_dates():
    curr_dir = os.getcwd()
    folder = 'fluorescence_outs'
    paths = {}
    for directory in os.listdir(folder):
        curr_path = os.path.join(folder, directory)
        
        date = directory.split('_')[0]
        for csv_file in os.listdir(curr_path):
            curr_path = os.path.join(curr_path, csv_file)
            paths[date] = curr_path
    print (paths)        
    return paths
```

```python
dates =  get_csv_files_paths_and_dates()
dataframes = []
for date in dates:
    dataframes.append(psii_complete.loc[psii_complete['date'] == date])

```

```python
combined = pd.concat(dataframes, ignore_index=True)
combined.to_csv('combined.csv')
display(combined)
```
