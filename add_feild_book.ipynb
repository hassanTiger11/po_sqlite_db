```python
'''
This table will store the meta information about each plant
'''
import sqlite3
sql_create_field_book = '''CREATE TABLE IF NOT EXISTS field_book (

plant_name text PRIMARY KEY NOT NULL UNIQUE,
treatment text NOT NULL,
plot text NOT NULL,
genotype text NOT NULL,
season integer DEFAULT 10,
species text DEFAULT "lettuce"
);
'''
```

```python
con = sqlite3.connect('po5.db')
cursor = con.cursor()
cursor.execute(sql_create_field_book)

```

```python
import pandas as pd
from IPython.display import display
'''
I nee to collect info from other tables about treatments and plots
I will use rgb to get this info
rgb[plant_name]'''
rgb = pd.read_sql_query("SELECT * FROM rgb", con)
display(rgb)
```

```python
def push_to_sql(plant_name, treatment, plot, genotype, season=10, species='lettuce'):
    sql = f'''INSERT INTO field_book(plant_name, treatment, plot, genotype, season, species) 
    VALUES("{plant_name}", "{treatment}", "{plot}", "{genotype}", {season}, "{species}");'''
    cur = con.cursor()
    
    cur.execute(sql, rgb)
    print("executed !")
    con.commit()
    print("commited !")
```

```python
for i in range(len(rgb)):
    if(i == 10): break
    plant_name= rgb.loc[i, 'plant_name']
    treatment= rgb.loc[i, 'treatment']
    plot= rgb.loc[i, 'plot']
    genotype= rgb.loc[i, 'genotype']
    season=10
    species='lettuce'
    print(f"push_to_sql({plant_name}, {treatment}, {plot}, {genotype}, season=10, species='lettuce')")
    push_to_sql(plant_name, treatment, plot, genotype, season, species)
    
    
```

```python
cols = ['plant_name', 'treatment', 'plot', 'genotype']
rgb[cols].to_sql('field_book', con, if_exists='append')
```
