```python
'''
In the following I will edit rgb table
I will drop all the non relevant columns like:
    treatment
    plot 
    genotype 
I will keep the plant_name so it can be a foriegn key to field_book
That way we can lose redundant data and we can get this data from the 
field_book.

One conclusion I got from field book:
      {treatment, plot, genotype} hasn't changed overtime for each scan
'''
import sqlite3
import pandas as pd
from IPython.display import display

# global values
po7 = sqlite3.connect('po7.db')
po8 = sqlite3.connect('po8.db')
scanner3d = pd.read_sql_query('SELECT * FROM scanner3d', po7)
```

```python
#inspect table
display(scanner3d)
```

```python
'''
ops:
1) drop columns
2) set primary keys
'''
#1) drop columns
drop_unwanted_columns = scanner3d.drop(columns=['index', 'Unnamed: 0'])
display(drop_unwanted_columns)
```

```python
'''
From the data we can see tha each plant has one scan per date
let's check that out
'''
#check that date+name is not redundant
name_date = drop_unwanted_columns['plant_name'] + drop_unwanted_columns['date']
print(f'len(name_date): {len(name_date)}')
print(f'len(set(name_date)): {len(set(name_date))}')
if(len(name_date) == len(set(name_date))):
    print('equal')
#Thus, the combination of date name is unique
```

```python
'''
commit new table drop_unwanted_columns
then edit with the script in the next cell
'''
schema = '''
CREATE TABLE IF NOT EXISTS "scanner3d" (
  "plant_name" TEXT NO NULL,
  "min_x" REAL,
  "min_y" REAL,
  "min_z" REAL,
  "max_x" REAL,
  "max_y" REAL,
  "max_z" REAL,
  "num_points" REAL,
  "hull_volume" REAL,
  "oriented_bounding_box" REAL,
  "axis_aligned_bounding_box" REAL,
  "persistence_entropy_0" REAL,
  "persistence_entropy_1" REAL,
  "persistence_entropy_2" REAL,
  "number_points_0" REAL,
  "number_points_1" REAL,
  "number_points_2" REAL,
  "amplitude_landscape_0" REAL,
  "amplitude_landscape_1" REAL,
  "amplitude_landscape_2" REAL,
  "amplitude_bottleneck_0" REAL,
  "amplitude_bottleneck_1" REAL,
  "amplitude_bottleneck_2" REAL,
  "amplitude_wasserstein_0" REAL,
  "amplitude_wasserstein_1" REAL,
  "amplitude_wasserstein_2" REAL,
  "amplitude_betti_0" REAL,
  "amplitude_betti_1" REAL,
  "amplitude_betti_2" REAL,
  "amplitude_silhouette_0" REAL,
  "amplitude_silhouette_1" REAL,
  "amplitude_silhouette_2" REAL,
  "amplitude_heat_0" REAL,
  "amplitude_heat_1" REAL,
  "amplitude_heat_2" REAL,
  "amplitude_persistence_image_0" REAL,
  "amplitude_persistence_image_1" REAL,
  "amplitude_persistence_image_2" REAL,
  "Unnamed: 0" REAL,
  "date" TEXT NO NULL,
  PRIMARY KEY ( "plant_name", "date")
);
'''
drop_unwanted_columns.to_sql('scanner3d', po8, if_exists='replace')
```

```python
sql_command = f'''

PRAGMA foreign_keys=off;

BEGIN TRANSACTION;

ALTER TABLE scanner3d RENAME TO old_scanner3d;

-- define the primary key constraint here
{schema}

INSERT INTO scanner3d SELECT * FROM old_scanner3d;

DROP TABLE old_scanner3d;

COMMIT;

PRAGMA foreign_keys=on;


'''
cursor = po8.cursor()
cursor.executescript(sql_command)
```
